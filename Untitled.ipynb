# With regex

```python
import re
```

```python
pattern = re.compile("Chapter [0-9]")
re.findall(pattern, book)
```

```python
pattern = re.compile("Chapter [0-9]+")
re.findall(pattern, book)
```

```python
len(re.findall(pattern, book))
```

# Which are the sentence where love was used ?

```python
pattern = re.compile("[^.]* love [^.]*.")
findings = re.findall(pattern, book)
len(findings)
```

### But "love," are not included so........

```python
pattern = re.compile("[^.]*[^a-zA-Z]+love[^a-zA-Z]+[^.]*.")
findings = re.findall(pattern, book)
len(findings)
```

```python
pattern = re.compile("[A_Z]{1}[^.]*[^a-zA-Z]+love[^a-zA-Z]+[^.]*.")
findings = re.findall(pattern, book)
len(findings)
```

# What are the most used words

```python
pattern = re.compile("[a-zA-Z]+")
finding = re.findall(pattern, book.lower())
finding[:5]
```

```python
d = {}
for word in finding:
    if word in d.keys():
        d[word] = d[word] + 1
    else:
        d[word] = 1
```

```python
d = {}
for word in finding:
    if word in d.keys():
        d[word] = d[word] + 1
    else:
        d[word] = 1
```
