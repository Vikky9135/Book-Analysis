# Loading the book

```python
with open("miracle_in_the_andes.txt", "r", encoding='UTF-8') as file:
    book = file.read()
```

# The most used words (Non-Articles) 

```python
import re
pattern = re.compile("[a-zA-Z]+")
finding = re.findall(pattern, book.lower())
finding[:5]
```

```python
d = {}
for word in finding:
    if word in d.keys():
        d[word] = d[word] + 1
    else:
        d[word] = 1
```

```python
d_list = [(value, key) for (key, value) in d.items()]
d_list = sorted(d_list, reverse=True)
d_list[:5]
```

```python
filtered_words = []
for count, word in d_list:
    if word not in english_stopword:
        filtered_words.append((word, count))
```

## Sentiment Analysis : What is the most positive and negetive chapter??

```python
import nltk
nltk.download('vader_lexicon')
```

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
analyzer 
```

```python
scores = analyzer.polarity_scores("Hey, I need you")
```

```python
scores
```

```python
if scores["pos"] >= scores["neg"]:
    print("Its a positive text")
else:
    print("Its a negetive text")
```

# Chapter sentiment analysis

```python
import re
pattern = re.compile("Chapter [0-9]+")
chapters = re.split(pattern, book)
```

```python
chapters = chapters[1:]
```

```python
for nr, chapter in enumerate(chapters):
    scores = analyzer.polarity_scores(chapter)
    print(nr+1, scores)
```
